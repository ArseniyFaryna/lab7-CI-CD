```python
%sql
CREATE OR REPLACE FUNCTION dbr_dev.faryna_gold.mask_load_date(load_date DATE)
RETURNS DATE
RETURN CASE
  WHEN current_user() = 'farynaarseniy2007@softserve.academy' THEN load_date 
  ELSE DATE('1900-01-01') 
END;
```

```python
%sql
ALTER MATERIALIZED VIEW dbr_dev.faryna_gold.agg_weather_daily
ALTER COLUMN load_date SET MASK dbr_dev.faryna_gold.mask_load_date;
```

```python
%sql
SELECT current_user();
SELECT * FROM dbr_dev.faryna_gold.agg_weather_daily;
```

```python
%sql
CREATE OR REPLACE FUNCTION dbr_dev.faryna_gold.weather_row_filter(load_date DATE)
RETURNS BOOLEAN
RETURN load_date >= '2026-03-15';
```

```python
%sql
CREATE OR REPLACE VIEW dbr_dev.faryna_gold.weather_secure_view AS
SELECT *
FROM dbr_dev.faryna_gold.agg_weather_daily
WHERE load_date IS NOT NULL 
  AND load_date >= '2026-03-15';
```
